# Pareto Analysis: Quality vs Cost

Which model dies first? Is Q3_K_M ever worth it?

This notebook reproduces the key findings from the benchmark suite.

```python
import pandas as pd
import matplotlib.pyplot as plt
import sys
sys.path.insert(0, '..')

from src.analysis.pareto import all_pareto_fronts

df = pd.read_parquet('../results/aggregated.parquet')
fronts = all_pareto_fronts(df)
fronts
```

```python
# Quality vs Latency Pareto
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for ax, (label, front) in zip(axes, fronts.items()):
    for model_id, group in front.groupby('model_id'):
        ax.plot(group.iloc[:, 0], group['quality_score'], marker='o', label=model_id)
    ax.set_title(label)
    ax.legend(fontsize=7)

plt.tight_layout()
plt.savefig('../reports/pareto_analysis.png', dpi=150)
plt.show()
```
